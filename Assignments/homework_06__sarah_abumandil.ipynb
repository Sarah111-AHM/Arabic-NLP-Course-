````python
import re
import random

import torch
import matplotlib.pyplot as plt
from datasets import Dataset
from transformers import AutoModelForCausalLM, AutoTokenizer
from trl import GRPOConfig, GRPOTrainer, DPOConfig, DPOTrainer
from peft import LoraConfig

random.seed(42)
torch.manual_seed(42)

MODEL = "Qwen/Qwen3-0.6B"
DIGITS = 3
````

## Utils

````python
def make_addition_dataset(n, digits=DIGITS, seed=0):
    rng = random.Random(seed)
    lo, hi = 10 ** (digits - 1), 10 ** digits - 1
    rows = []
    for _ in range(n):
        a, b, c = (rng.randint(lo, hi) for _ in range(3))
        rows.append({"question": f"What is {a} + {b} + {c}?", "answer": str(a + b + c)})
    return Dataset.from_list(rows)
````

````python
ANSWER_RE = re.compile(r"<answer>\s*(-?\d+)\s*</answer>")

def extract_answer(text: str) -> str | None:
    m = ANSWER_RE.search(text)
    return m.group(1) if m else None
````

````python
def plot_reward(log_history, key="reward", title=None):
    points = [(e["step"], e[key]) for e in log_history if key in e]
    if not points:
        print(f"No '{key}' logged. Available:", sorted({k for e in log_history for k in e}))
        return
    steps, vals = zip(*points)
    plt.figure(figsize=(6, 3))
    plt.plot(steps, vals)
    plt.xlabel("step"); plt.ylabel(key); plt.title(title or key)
    plt.grid(alpha=0.3); plt.show()
````

````python
@torch.no_grad()
def accuracy(model, tokenizer, dataset, max_new_tokens=96):
    model.eval()
    correct = 0
    for row in dataset:
        inputs = tokenizer(row["prompt"], return_tensors="pt").to(model.device)
        out = model.generate(**inputs, max_new_tokens=max_new_tokens, do_sample=False)
        text = tokenizer.decode(out[0, inputs.input_ids.shape[1]:], skip_special_tokens=True)
        correct += int(extract_answer(text) == row["answer"])
    return correct / len(dataset)
````

## Part 1 - Baseline

````python
tokenizer = AutoTokenizer.from_pretrained(MODEL)
model = AutoModelForCausalLM.from_pretrained(MODEL, torch_dtype=torch.float16, device_map="auto")

def add_prompt(row):
    messages = [{"role": "user",
                 "content": row["question"] + " Put your final answer inside <answer>...</answer> tags."}]
    text = tokenizer.apply_chat_template(
        messages, tokenize=False, add_generation_prompt=True, enable_thinking=False
    )
    return {"prompt": text}

train_ds = make_addition_dataset(200, seed=0).map(add_prompt)
eval_ds  = make_addition_dataset(50,  seed=1).map(add_prompt)
````

````python
baseline_acc = accuracy(model, tokenizer, eval_ds)
print(f"Baseline accuracy: {baseline_acc:.1%}")

sample = eval_ds[0]
inputs = tokenizer(sample["prompt"], return_tensors="pt").to(model.device)
out = model.generate(**inputs, max_new_tokens=96, do_sample=False)
print("\nQ:", sample["question"])
print("Model:", tokenizer.decode(out[0, inputs.input_ids.shape[1]:], skip_special_tokens=True))
print("Truth:", sample["answer"])
````

## Part 2 - Reward design

A GRPO reward function has the signature:

```python
def my_reward(completions, **kwargs) -> list[float]:
    ...
```

`completions` is a list of strings (one per sampled generation from our model). Any other dataset column (e.g. `answer`) is forwarded through `kwargs`.

### Exercise 1 - Format reward

Return **1.0** when the completion contains `<answer>NUMBER</answer>`, **0.0** otherwise.

````python
def format_reward(completions, **kwargs):
    return [1.0 if extract_answer(c) is not None else 0.0 for c in completions]
````

### Exercise 2 - Correctness reward

Return **1.0** when the extracted answer equals the ground truth, **0.0** otherwise. Ground truths arrive as `kwargs["answer"]`, a list aligned with `completions`.

````python
def correctness_reward(completions, **kwargs):
    truths = kwargs["answer"]
    return [1.0 if extract_answer(c) == gt else 0.0 for c, gt in zip(completions, truths)]
````

## Part 3 - Train with GRPO

````python
lora = LoraConfig(
    r=16, lora_alpha=32, task_type="CAUSAL_LM",
    target_modules=["q_proj", "k_proj", "v_proj", "o_proj"],
)

grpo_cfg = GRPOConfig(
    output_dir="grpo-out",
    learning_rate=5e-6,
    per_device_train_batch_size=4,
    num_generations=4,
    max_completion_length=96,
    max_steps=80,
    logging_steps=5,
    save_strategy="no",
    fp16=True,
    beta=0.04,
    report_to="none",
)

trainer = GRPOTrainer(
    model=model,
    reward_funcs=[format_reward, correctness_reward],
    args=grpo_cfg,
    train_dataset=train_ds,
    peft_config=lora,
    processing_class=tokenizer,
)
trainer.train()
````

## Part 4 - Did it work?

````python
grpo_acc = accuracy(trainer.model, tokenizer, eval_ds)
print(f"Baseline:   {baseline_acc:.1%}")
print(f"After GRPO: {grpo_acc:.1%}")
plot_reward(trainer.state.log_history, key="reward", title="GRPO reward (format + correctness)")
````

## Part 5 - Goodhart (required)

> *"When a measure becomes a target, it ceases to be a good measure."*

Drop correctness, keep format. Watch reward climb while accuracy does not.

````python
model_hack = AutoModelForCausalLM.from_pretrained(MODEL, torch_dtype=torch.float16, device_map="auto")

grpo_hack_cfg = GRPOConfig(
    output_dir="grpo-hack",
    learning_rate=5e-6,
    per_device_train_batch_size=4,
    num_generations=4,
    max_completion_length=96,
    max_steps=80,
    logging_steps=5,
    save_strategy="no",
    fp16=True,
    beta=0.04,
    report_to="none",
)

trainer_hack = GRPOTrainer(
    model=model_hack,
    reward_funcs=[format_reward],
    args=grpo_hack_cfg,
    train_dataset=train_ds,
    peft_config=lora,
    processing_class=tokenizer,
)
trainer_hack.train()

hack_acc = accuracy(trainer_hack.model, tokenizer, eval_ds)
print(f"Baseline:              {baseline_acc:.1%}")
print(f"Format-only training:  {hack_acc:.1%}")
plot_reward(trainer_hack.state.log_history, key="reward", title="Format-only reward (hacked)")
````

### What you should see

- Reward climbs toward its ceiling.
- Accuracy hovers near baseline or drifts down.

The fix is not a cleverer parser. It is to design rewards that cannot be satisfied without solving the task.

## Part 6 (Optional) - DPO

DPO skips sampling: give it `(chosen, rejected)` pairs and it pushes the policy toward the chosen ones directly. We build pairs from the baseline model's own rollouts.

````python
@torch.no_grad()
def build_pairs(model, tokenizer, dataset, k=4, max_new_tokens=96):
    model.eval()
    pairs = []
    for row in dataset:
        inputs = tokenizer(row["prompt"], return_tensors="pt").to(model.device)
        outs = model.generate(**inputs, max_new_tokens=max_new_tokens,
                              do_sample=True, temperature=1.0, num_return_sequences=k)
        cont = inputs.input_ids.shape[1]
        texts = [tokenizer.decode(o[cont:], skip_special_tokens=True) for o in outs]
        good = next((t for t in texts if extract_answer(t) == row["answer"]), None)
        bad  = next((t for t in texts if extract_answer(t) != row["answer"]), None)
        if good and bad:
            pairs.append({"prompt": row["prompt"], "chosen": good, "rejected": bad})
    return Dataset.from_list(pairs)

base = AutoModelForCausalLM.from_pretrained(MODEL, torch_dtype=torch.float16, device_map="auto")
pref_ds = build_pairs(base, tokenizer, train_ds.select(range(60)))
print(f"Built {len(pref_ds)} preference pairs")
````

````python
model_dpo = AutoModelForCausalLM.from_pretrained(MODEL, torch_dtype=torch.float16, device_map="auto")

dpo_cfg = DPOConfig(
    output_dir="dpo-out",
    learning_rate=5e-6,
    per_device_train_batch_size=4,
    max_length=512,
    num_train_epochs=1,
    logging_steps=5,
    save_strategy="no",
    fp16=True,
    beta=0.1,
    report_to="none",
)

trainer_dpo = DPOTrainer(
    model=model_dpo,
    args=dpo_cfg,
    train_dataset=pref_ds,
    peft_config=lora,
    processing_class=tokenizer,
)
trainer_dpo.train()

dpo_acc = accuracy(trainer_dpo.model, tokenizer, eval_ds)
print(f"Baseline:   {baseline_acc:.1%}")
print(f"After GRPO: {grpo_acc:.1%}")
print(f"After DPO:  {dpo_acc:.1%}")
````
